```python
import nltk
import string
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.corpus import brown
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
#Reading Book
file = open("book2.txt",encoding="utf8")

# variable T
T = file.read()

# variable T2
T2=T

# Printing the contents of the book stored in T
print(T[:1000])
```

```python
translator = str.maketrans("","",string.punctuation)
T=T.translate(translator)
print(T[:1000])
```

```python
# Removing acknowledgement
T = re.sub("Project[\s\S]*CONTENTS","",T)
T = re.sub("TRANSCRIBER[\s\S]*","",T)
print(T[:1000])
```

```python
# Removing Chapter Names
T=re.sub("[A-Z]{2,}","",T)
print(T[:1000])
```

```python
# Removing Chapter Numbers
T=re.sub("[0-9]+","",T)
print(T)
```

```python
# Converting text to lowercase
T = T.lower()
print(T[:1000])
```

```python
# splitting the lines and joining them into one.
T = " ".join(T.split())
wordcloudT=T
T2=T
print(T[:1000])
```
